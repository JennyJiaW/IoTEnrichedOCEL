#### Flattern Truck

```python
truck_df = pd.read_csv('object_Truck.csv')
truck_df = truck_df['ocel_id;ocel_time;ocel_changed_field;RFID No;Pickup Plan ID;Cargo ID;LPT;Axles;Scheduled Pickup Weight;Truck Status;Truck Weight;Is_normal'].str.split(';', expand=True)
truck_df.columns = ['ocel_id','ocel_time','ocel_changed_field','RFID No','Pickup Plan ID','Cargo ID','LPT','Axles','Scheduled Pickup Weight','Truck Status','Truck Weight','Is_normal']
truck_df.rename(columns={'ocel_id': 'ocel_object_id', 'ocel_time': 'timestamp'}, inplace=True)
```

```python
truck_df
```

```python
trucks_merged_df = pd.merge(events_distinct, truck_df, on=['ocel_object_id', 'timestamp'], how='outer')
trucks_merged_df
```

```python
truck_filtered_df=trucks_merged_df[trucks_merged_df['ocel_object_id'].str.startswith('tr')]
truck_filtered_df
```

```python
# Prior to the pick-up process, the truck's data (e.g. RFID number, LPT, axle number) is stored in the port system.
truck_filtered_df_cleaned = truck_filtered_df.dropna(subset=['event_id'])
truck_filtered_df_cleaned
```

```python
assign_trucks_df = truck_filtered_df_cleaned[truck_filtered_df_cleaned['activity'] == 'Assign Trucks']
assign_trucks_df
```

```python
other_activities_df = truck_filtered_df_cleaned[truck_filtered_df_cleaned['activity'] != 'Assign Trucks']
other_activities_df
```

```python
def merge_group_rows_with_field(group):
    merged_row = group.iloc[0].copy()
    for col in group.columns:
        if col not in ['event_id', 'activity', 'timestamp', 'ocel_object_id', 'ocel_IoT_object_id', 'ocel_changed_field']:
            non_null_values = group[col].dropna().unique()
            if len(non_null_values) > 0:
                merged_row[col] = non_null_values[0]

    # Populate the value from 'ocel_changed_field' into the corresponding column
    for _, row in group.iterrows():
        field = row['ocel_changed_field']
        if field in group.columns and pd.notna(row[field]):
            merged_row[field] = row[field]

    return merged_row
```

```python
# Group by 'event_id' and 'timestamp' and merge rows within each group considering 'ocel_changed_field'
assigned = assign_trucks_df.groupby(['event_id', 'timestamp']).apply(merge_group_rows_with_field).reset_index(drop=True)
```

```python
assigned
```

```python
trucks_xes = pd.concat([assigned, other_activities_df]).sort_values(by='timestamp').reset_index(drop=True)
```

```python
trucks_xes 
```

```python
fail_df = trucks_xes[trucks_xes ['activity'] == 'Fail to Weigh']
fail_df
```

```python
rest_df = trucks_xes[trucks_xes ['activity'] != 'Fail to Weigh']
rest_df
```

```python
fail = fail_df.groupby(['event_id', 'timestamp']).apply(merge_group_rows_with_field).reset_index(drop=True)
fail
```

```python
trucks = pd.concat([fail, rest_df]).sort_values(by='timestamp').reset_index(drop=True)
trucks
```

```python
# Order the dataframe by 'ocel_object_id' and 'timestamp'
truck_ordered = trucks.sort_values(by=['ocel_object_id', 'timestamp']).reset_index(drop=True)

# Reinitialize the 'pickup' column
truck_ordered['pickup'] = None

# Initialize variables to track pickups
pickup_id = 0
assigning_pickup = False
current_truck = None

# Iterate over the rows to assign pickup IDs
for idx, row in truck_ordered.iterrows():
    activity = row['activity']
    truck_id = row['ocel_object_id']

    if activity == "Assign Trucks":
        pickup_id += 1
        assigning_pickup = True
        current_truck = truck_id
        truck_ordered.at[idx, 'pickup'] = pickup_id

    elif assigning_pickup and truck_id == current_truck:
        truck_ordered.at[idx, 'pickup'] = pickup_id
        if activity in ["Evaluate the truck exit", "fail to weigh"]:
            assigning_pickup = False
```

```python
truck_ordered
```

```python
pickup_plan_ids = truck_ordered.dropna(subset=['Pickup Plan ID']).drop_duplicates(subset=['pickup'])[['pickup', 'Pickup Plan ID']]
pickup_plan_ids
```

```python
data = pd.merge(truck_ordered, pickup_plan_ids, on='pickup', how='left', suffixes=('', '_filled'))
data
```

```python
data = data.drop(columns=['ocel_changed_field','Pickup Plan ID'])
data
```

```python
data.rename(columns={'Pickup Plan ID_filled': 'Pickup Plan ID'}, inplace=True)
```
